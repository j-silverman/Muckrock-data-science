```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

dfGun = pd.read_csv('gunOwnership.csv')
dfTweet = pd.read_csv('stateCount.csv')

gunF = np.asarray(dfGun['Percentage'])
tweetF = np.asarray(dfTweet['frequency'])
```

```python
fig=plt.figure(figsize=(20, 14), dpi= 80, facecolor='w', edgecolor='k')

_ = plt.scatter(gunF,tweetF,s=200,label='State')
_ = plt.xlabel('Gun Ownership Percentage by State',fontsize = 20)
_ = plt.ylabel('Frequency of Tweets per Congresspeople per State',fontsize = 20)

x = np.vstack([gunF,np.ones(len(gunF))]).T

m,c = np.linalg.lstsq(x,tweetF,rcond=-1)[0]
coef = np.linalg.lstsq(x,tweetF,rcond=-1)[3]

print(m,c)
_=plt.plot(gunF,m*gunF+c,'r',label='Regression',linewidth= 5)
_=plt.title('Linear Regression using Least Squares',fontsize = 30)
_=plt.annotate('$R^2 = 0.3051$',xy = [53.1,70],xytext=[53.7,80],fontsize=30)
_=plt.annotate('$B_0 = 53.03$',xy = [53.1,65],xytext=[53.7,75],fontsize=30)
_=plt.annotate('$B_1 = -0.882$',xy = [53.1,60],xytext=[53.7,70],fontsize=30)

_=plt.legend(prop={'size': 20})
plt.show()
```
